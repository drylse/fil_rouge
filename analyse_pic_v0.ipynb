```python
data_new2=data_new[data_new.Bande!='XXX']
# print(data_new2)
data_new3=data_new2.dropna(subset=['N2O gN.ha-1.j-1'])
pd.DataFrame(data=data_new3)
```

```python
data_new4 = data_new3.loc[~((pd.isna(data_new2['%H - h1']) & (pd.isna(data_new2['%H - h2'])) 
                            & (pd.isna(data_new2['NO3 Sol - h1\nkgN.ha-1'])) & (pd.isna(data_new2["NO3 Sol - h2\nkgN.ha-1"]))
                            & (pd.isna(data_new2['NH4 Sol - h1\nkgN.ha-1'])) & (pd.isna(data_new2['NH4 Sol - h2\nkgN.ha-1']))
                            & (pd.isna(data_new2['Fertilisation (kgN.ha-1)'])) & (pd.isna(data_new2['N biomasse aérienne\n (KgN.ha-1)']))
                            & (pd.isna(data_new2['INN ou N Tester'])) & (pd.isna(data_new2['INN ou N Tester']))
                            & (pd.isna(data_new2['PRS N-NO3\n(µg.10cm-2.14j-1)'])) & (pd.isna(data_new2['PRS N-NH4\n(µg.10cm-2.14j-1)']))
                            & (pd.isna(data_new2['PRS N-NH4\n(µg.10cm-2.14j-1)'])) & (pd.isna(data_new2['PRS N-NH4 ZéroN\n(µg.10cm-2.14j-1)']))
                            ))]
```
