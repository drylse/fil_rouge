```python
# Importer les librairies nécessaires
import Orange
from Orange.data import Table

# Créer le dataframe
data = [["A", "B", "C", "D", "E"],
        ["A", "B", "D", "E"],
        ["B", "C", "D", "E"],
        ["A", "B", "C", "D", "E"],
        ["A", "B", "C", "E"]]

# Créer un tableau Orange
table = Table(data)

# Détecter les motifs fréquents
frequent_itemsets = Orange.associate.AssociationRulesSparseInducer(table, support = 0.6).get_itemsets(0.6)

# Afficher les résultats
print(frequent_itemsets)
```
